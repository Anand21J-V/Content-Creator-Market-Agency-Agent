```python
import os
from dotenv import load_dotenv

load_dotenv()
groq_api_key = os.getenv("GROQ_API_KEY")

if not groq_api_key:
    raise ValueError("Missing GROQ_API_KEY in .env")
```

```python
from groq import Groq

groq_llm = Groq(api_key=groq_api_key)
```

```python
from langgraph.graph import StateGraph, END
from langchain_core.runnables import RunnableLambda

# Shared state: input prompt and outputs
state = {
    "topic": str,
    "trends": str,
    "competitor_analysis": str,
    "content_ideas": str,
}

# Step 1: Research trends
def trend_researcher_node(state):
    topic = state["topic"]
    prompt = f"What are the latest digital content trends around: {topic}?"
    response = groq_llm.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="llama-3.3-70b-versatile"
    )
    return {**state, "trends": response.choices[0].message.content}


def competitor_analyst_node(state):
    topic = state["topic"]
    prompt = f"Do a quick competitor content analysis around this niche: {topic}. What are others doing successfully?"
    response = groq_llm.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="llama-3.3-70b-versatile"
    )
    return {**state, "competitor_analysis": response.choices[0].message.content}


# Step 3: Content Idea Generation
def idea_generator_node(state):
    topic = state["topic"]
    trends = state["trends"]
    competitors = state["competitor_analysis"]
    prompt = f"""Given the topic "{topic}", trends: {trends}, and competitor activity: {competitors},
    generate 5 unique, viral-ready content ideas suitable for social media, blogs, or email campaigns."""
    response = groq_llm.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="llama-3.3-70b-versatile"
    )
    return {**state, "content_ideas": response.choices[0].message.content}

```

```python
graph = StateGraph(state)

graph.add_node("TrendResearcher", RunnableLambda(trend_researcher_node))
graph.add_node("CompetitorAnalyst", RunnableLambda(competitor_analyst_node))
graph.add_node("IdeaGenerator", RunnableLambda(idea_generator_node))

# Chain connections
graph.set_entry_point("TrendResearcher")
graph.add_edge("TrendResearcher", "CompetitorAnalyst")
graph.add_edge("CompetitorAnalyst", "IdeaGenerator")
graph.add_edge("IdeaGenerator", END)

# Compile graph
content_graph = graph.compile()
```

```python
topic = "Sustainable fashion marketing"

result = content_graph.invoke({"topic": topic})

from pprint import pprint
pprint(result)
```
